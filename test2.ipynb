```python
X = [['Area_sqft', 'Bedrooms', 'Distance_to_city_km']]
y = ['Price']
```

```python
df_filled = df.fillna(df.mean())
X = df_filled[['Area_sqft', 'Bedrooms', 'Distance_to_city_km']]
y = df_filled['Price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
print("X shape:", X.shape)
print("y shape:", y.shape)
print("Any NaNs?", X.isna().sum().sum() > 0 or y.isna().sum() > 0)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R Score: {r2:.4f}")
```

```python
import pandas as pd
import numpy as np
import random

np.random.seed(42)
random.seed(42)

# Create a larger, realistic dataset (200 rows)
rows = 200

data = {
    "Area_sqft": np.random.randint(500, 4000, size=rows).astype(object),
    "Bedrooms": np.random.choice([1, 2, 3, 4, 5, 'three', np.nan], size=rows),
    "Distance_to_city_km": np.random.choice(
        [round(np.random.uniform(2, 30), 1) for _ in range(100)] + ['10km', np.nan, 'fifteen'], size=rows
    ),
    "Age_of_house_yrs": np.random.choice([0, 1, 2, 5, 10, 15, 20, 30, np.nan], size=rows),
    "Has_Garage": np.random.choice(['Yes', 'No', 'Y', 'N', np.nan, 'yes', 'no'], size=rows),
    "Price": np.random.randint(50000, 800000, size=rows).astype(object),
}

df_dirty = pd.DataFrame(data)

# Introduce some dirty issues
for _ in range(20):
    i = np.random.randint(0, len(df_dirty))
    df_dirty.loc[i, "Area_sqft"] = str(df_dirty.loc[i, "Area_sqft"]) + " sqft"

for _ in range(10):
    i = np.random.randint(0, len(df_dirty))
    df_dirty.loc[i, "Price"] = "N/A"

# Add a few duplicates
df_dirty = pd.concat([df_dirty, df_dirty.sample(5)], ignore_index=True)

# Add some leading/trailing spaces
df_dirty["Has_Garage"] = df_dirty["Has_Garage"].apply(lambda x: f" {x} " if isinstance(x, str) else x)

# Save to CSV
csv_path = "dirty_housing_data.csv"
df_dirty.to_csv(csv_path, index=False)

print(f" Dirty dataset generated and saved as '{csv_path}'")
print("Shape:", df_dirty.shape)
print("\nSample:")
print(df_dirty.head(10))
```
